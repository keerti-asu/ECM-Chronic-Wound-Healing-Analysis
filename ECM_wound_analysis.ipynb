# ECM Chronic Wound Healing Analysis
Exploratory analysis of ECM clinical trial outcomes.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```
df = pd.read_csv('https://raw.githubusercontent.com/keerti-asu/ECM-Chronic-Wound-Healing-Analysis/refs/heads/main/clinical_trials_ecm.csv')
df.head()
```

## Healing Rate Comparison

```
df[['treatment_healing_rate','control_healing_rate']].mean()
```

## Boxplot: Healing Time

```
sns.boxplot(data=df[['treatment_time_days','control_time_days']])
plt.title('Healing Time Comparison')
plt.show()
```

## Correlation Between Healing Rate and Time

```
corr = df['treatment_healing_rate'].corr(df['treatment_time_days'])
corr
```

## Statistical Test

```
stats.ttest_ind(df['treatment_time_days'], df['control_time_days'])
```
